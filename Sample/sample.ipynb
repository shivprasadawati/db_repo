```python
from pyspark.sql.functions import to_date
from pyspark.sql.functions import *

# rename columns

df = spark.read.format("csv").option("header", "true").load("/Volumes/workspace/default/healthcare")
df = df.withColumnRenamed('Blood Type', 'BloodType').withColumnRenamed('Medical Condition', 'MedicalCondition').withColumnRenamed('Date of Admission', 'DateofAdmission').withColumnRenamed('Insurance Provider', 'InsuranceProvider').withColumnRenamed('Billing Amount', 'BillingAmount').withColumnRenamed('Room Number', 'RoomNumber').withColumnRenamed('Admission Type', 'AdmissionType').withColumnRenamed('Discharge Date', 'DischargeDate').withColumnRenamed('Test Results', 'TestResults')


# casted with suitable format 

df = df.withColumn("Age", df["Age"].cast("int")).withColumn("BillingAmount", df["BillingAmount"].cast("double")).withColumn("RoomNumber", df["RoomNumber"].cast("int")).withColumn("DateofAdmission", to_date(df["DateofAdmission"], "yyyy-MM-dd")).withColumn("DischargeDate", to_date(df["DischargeDate"], "yyyy-MM-dd"))

df.display()

# loaded into the table
df.write.format("delta").mode("overwrite").saveAsTable("healthcare")
```
